```
from os import listdir
from os.path import isfile, join
mypath='/gdrive/My Drive/AI_Core/_data/roses/'
destpath='/gdrive/My Drive/AI_Core/_data/roses_dest/'
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]


for f in onlyfiles:
  img = cv2.imread(join(mypath,f), cv2.IMREAD_UNCHANGED)
  dst1 = cv2.resize(img, dsize=(200, 140))
  dst2 = cv2.cvtColor(dst1, cv2.COLOR_BGR2GRAY)
  fPath=join(destpath,f)
  cv2.imwrite(fPath,dst2)
  cv2_imshow(dst2)
```
